```python
plt.figure(figsize=(15, 10))
plt.plot(df_analysis2['Ratio'],df_analysis2['Total Complaints'], color='red')
plt.xlabel('Complaint Ratio', fontsize=25)
plt.ylabel('Number of Complaints',fontsize=25)
plt.title('Total Complaints',fontsize=25)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot(df_analysis2['Ratio'],df_analysis2['Upheld Complaints'],color='blue')
plt.xlabel('Complaint Ratio', fontsize=25)
plt.ylabel('Number of Complaints',fontsize=25)
plt.title('Upheld Complaints',fontsize=25)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot(df_analysis2['Ratio'],df_analysis2['Question of Fact Complaints'],color='k')
plt.xlabel('Complaint Ratio', fontsize=25)
plt.ylabel('Number of Complaints',fontsize=25)
plt.title('Question of Fact Complaints',fontsize=25)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot(df_analysis2['Ratio'],df_analysis2['Not Upheld Complaints'],color='green')
plt.xlabel('Complaint Ratio', fontsize=25)
plt.ylabel('Number of Complaints',fontsize=25)
plt.title('Not Upheld Complaints',fontsize=25)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot(df_analysis2['Ratio'],df_analysis2['Upheld Complaints'],color='blue')
plt.plot(df_analysis2['Ratio'],df_analysis2['Question of Fact Complaints'],color='k')
plt.plot(df_analysis2['Ratio'],df_analysis2['Not Upheld Complaints'],color='green')
plt.legend(loc='upper right',fontsize=15)
plt.xlabel('Complaint Ratio', fontsize=25)
plt.ylabel('Number of Complaints',fontsize=25)
plt.title('Sum of "Upheld Complaints", "Question of Fact Complaints", and "Not Upheld Complaints"',fontsize=25)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.grid(c = 'orange')
plt.show()
```
